```python
import xlrd
import os
import os.path
import nltk
from nltk import FreqDist
from nltk import bigrams
from nltk import ngrams
import gensim
import pyLDAvis
import pyLDAvis.gensim
import pandas

from gensim import corpora, models
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

pyLDAvis.enable_notebook()
```

```python
stops = set(stopwords.words("english"))
```

```python
fullText = []
for dirpath, dirnames, filenames in os.walk('/home/neil/Downloads'):
    for filename in [f for f in filenames if f.endswith(".xlsx")]:
        path = (os.path.join(dirpath, filename))
        book = xlrd.open_workbook(path)
        sheet = book.sheet_by_index(0)
        vals = sheet.col_values(1)
        fullText.extend(vals)
        print(path + " added")
print(fullText)
```

```python
texts = [[word for word in text.lower().split() if word not in stops] for text in fullText]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
print(texts)
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=10)
```

```python
Viewmodel = pyLDAvis.gensim.prepare(lda,corpus,dictionary)
pyLDAvis.save_html(Viewmodel,'lda.html')
```

```python
string = ' '.join(fullText).replace(".", "").replace(",", "").replace("\n", "").lower()
list2=string.split()
filtered =list(filter(lambda x: x not in stops,list2))
words = ' '.join(filtered)

tokens = nltk.word_tokenize(words)
text = nltk.Text(tokens)

```

```python

n = 3
bgs = ngrams(tokens, n)
fdist1 = FreqDist(bgs)
```
